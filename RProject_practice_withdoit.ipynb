### (실습과제) P2-1
* (가) 10대, 20대, ...,60대, 기타 연령대로 나누고 가장 수입을 많이 받는 연령대는?
* (나) 10대, 20대, ...,60대, 기타 연령대로 나누고 가장 수입을 적게 받는 연령대는?

```r
library(foreign)
library(dplyr)
library(ggplot2)
library(readxl)
```

```r
dat_welfare <- read.spss(file="D:\\dataset\\R_Doit\\Koweps_hpc10_2015_beta1.sav", to.data.frame=T)
welfare1 <- dat_welfare
```

```r
welfare <- rename(welfare1, 
                 sex=h10_g3,
                 birth=h10_g4,
                 marriage=h10_g10,
                 religion=h10_g11,
                 income=p1002_8aq1,
                 code_job=h10_eco9,
                 code_region=h10_reg7)
names(welfare)
```

```r
welfare <- select(welfare, sex, birth, marriage, religion, income, code_job, code_region)
names(welfare)
```

```r
welfare$age <- 2015 - welfare$birth + 1
welfare
```

```r
age_income <- welfare %>% 
              filter(!is.na(income)) %>%
              group_by(age) %>%
              summarise(mean_income = mean(income))
head(age_income)
```

```r
welfare <- welfare %>%
           mutate(class=ifelse(age<20, '10대',
                               ifelse(age<30, '20대',
                                 ifelse(age<40, '30대',
                                   ifelse(age<50, '40대',
                                     ifelse(age<60, '50대', 
                                      ifelse(age<70, '60대', "기타연령")))))))
table(welfare$class)
```

```r
ggplot(welfare, aes(x=class)) + geom_bar()
```

```r
### 연령대별 월급 평균표 만들기 
class_income <- welfare %>% filter(!is.na(income)) %>%
                          group_by(class) %>%
                          summarise(mean_income = mean(income))
class_income
```

### (실습해보기) P3-2
* (가) '성별로 어떤 직업이 많은지'에 대해 분석해보자.
* (Hint) summarise(n=n()) 은 해당 그룹에 대한 개수를 출력할 수 있다.

```r
job_male <- welfare %>%
        filter(!is.na(job) & sex=='male') %>%
        group_by(job) %>%
        summarise(n=n()) %>%
        arrange(desc(n)) %>%
        head(10)
job_male
```

```r
job_female <- welfare %>%
        filter(!is.na(job) & sex=='female') %>%
        group_by(job) %>%
        summarise(n=n()) %>%
        arrange(desc(n)) %>%
        head(10)
job_female
```

```r
ggplot(data = job_male, aes(x=reorder(job, n), y= n)) + 
           geom_col() +
           coord_flip()
```

```r
ggplot(data = job_female, aes(x=reorder(job, n), y= n)) + 
           geom_col() +
           coord_flip()
```
